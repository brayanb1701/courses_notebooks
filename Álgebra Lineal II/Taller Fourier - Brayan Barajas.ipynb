Considerando el espacio $C^0[-\pi,\pi]$ con el producto interno: <br>
<center>$ <f,g>=\int_{-\pi}^{\pi}f(x)g(x)dx$</center><br>
Considerando $a_{0}=6, a_{1}=8, a_{2}=8$, la función $f:[-\pi,\pi]\rightarrow {\rm I \!R}$ se define así:
$f(x)= \left\{ \begin{array}{lcc}
             18-x^2 &   si  & x < 0 \\
             \\ 8x+4 &  si &  x \geq 0 \\
             \end{array}
   \right.$ <br>
En código, esto se representaría de la siguiente forma:

```python
f=piecewise([[[-pi,0],18-x^2],[(0,pi),8*x+4]])
figf=plot(f,(-pi,pi),title="Funcion f(x) dada por el código")
(figf).show()
```

<strong>3. </strong>Encuentre la función de $W$ más cercana a  $f$ cuando $W$ es igual a cada uno de los generados dados.

<p align="justify">Para hacer los cálculos más rápidos, se creará una función que reciba el conjunto generador de funciones para realizar la proyección respectiva de la función $f(x)$.</p>

```python
def calc_proy(gen):
    #Se recorre cada una de las funciones del conjunto generador
    resultado=0
    for g in gen:
        #(<f,gi>/<gi,gi>)*gi donde gi es cada una de las funciones del conjunto generador
        ng=integral(g^2,x,-pi,pi) #norma de la función gi
        #producto interno coeficiente de Fourier
        pig=(1/ng)*(integral((18-x^2)*g,x,-pi,0)+integral((8*x+4)*g,x,0,pi))
        fg=pig*g
        #Se van sumando las funciones
        resultado=resultado+fg
    return resultado
```

<p align="justify">Para observar los diferentes resultados obtenidos con cada uno de los conjuntos generadores, se mostrará gráficamente el resultado tanto de la función creada <i>calc_proy()</i> como con <i>fourier_series_cosine_coefficient()</i> y <i>fourier_series_sin_coefficient()</i>. Adicionalmente, en el $(e)$ se utilizará <i>fourier_series_partial_sum()</i> ya que se tienen tanto senos como cosenos.
<br>
<b>Nota:</b> En el caso del coeficiente $c_0$, se debe utilizar <i>fourier_series_cosine_coefficient()</i> y el resultado obtenido debe dividirse por 2 ya que así está estipulado en la documentación de Sage.</p>

---
$(a) \: span\{1, sin(x)\}$

```python
#Se utiliza la función para calcular con el conjunto {1,sin(x)}
a=calc_proy([1,sin(x)])
#Se grafica el resultado obtenido
figa=plot(a,-pi,pi,color="red", title="Proyección con calc_proy()",legend_label='(a) calc_proy()')

#se utiliza fourier_series_cosine_coefficient() y fourier_series_sin_coefficient() como alternativa para comprobar.
cc0=f.fourier_series_cosine_coefficient(0)
cs1=f.fourier_series_sine_coefficient(1)
a1=(cc0/2)+cs1*sin(x)
#Se grafica el resultado obtenido
figa1=plot(a1,-pi,pi,color="orange",title="Proyección con coeficientes de fourier",legend_label='(a) coeficientes de fourier')

#Se utiliza graphics_array() para mostrar en dos gráfica los resultados anteriores
figure= graphics_array([figf+figa,figf+figa1])
figure.show(figsize=[8,4])
```

---
$(b) \: span\{1, sin(x), sin(2x)\}$

```python
#Se utiliza la función para calcular con el conjunto {1,sin(x),sin(2*x)}
b=calc_proy([1,sin(x),sin(2*x)])
#Se grafica el resultado obtenido
figb=plot(b,-pi,pi,color="red", title="Proyección con calc_proy()",legend_label='(b) calc_proy()')

#se utiliza fourier_series_cosine_coefficient() y fourier_series_sin_coefficient() como alternativa para comprobar.
cc0=f.fourier_series_cosine_coefficient(0)
cs1=f.fourier_series_sine_coefficient(1)
cs2=f.fourier_series_sine_coefficient(2)
b1=(cc0/2)+cs1*sin(x)+cs2*sin(2*x)
#Se grafica el resultado obtenido
figb1=plot(b1,-pi,pi,color="orange",title="Proyección con coeficientes de fourier",legend_label='(b) coeficientes de fourier')

#Se utiliza graphics_array() para mostrar en dos gráfica los resultados anteriores
figure= graphics_array([figf+figb,figf+figb1])
figure.show(figsize=[8,4])
```

---
$(c) \: span\{1, cos(x)\}$

```python
#Se utiliza la función para calcular con el conjunto {1,cos(x)}
c=calc_proy([1,cos(x)])
#Se grafica el resultado obtenido
figc=plot(c,-pi,pi,color="red", title="Proyección con calc_proy()",legend_label='(c) calc_proy()')

#se utiliza fourier_series_cosine_coefficient() y fourier_series_sin_coefficient() como alternativa para comprobar.
cc0=f.fourier_series_cosine_coefficient(0)
cc1=f.fourier_series_cosine_coefficient(1)
c1=(cc0/2)+cc1*cos(x)
#Se grafica el resultado obtenido
figc1=plot(c1,-pi,pi,color="orange",title="Proyección con coeficientes de fourier",legend_label='(c) coeficientes de fourier')

#Se utiliza graphics_array() para mostrar en dos gráfica los resultados anteriores
figure= graphics_array([figf+figc,figf+figc1])
figure.show(figsize=[8,4])
```

---
$(d) \: span\{1, cos(x), cos(2x)\}$

```python
#Se utiliza la función para calcular con el conjunto {1,cos(x),cos(2*x)}
d=calc_proy([1,cos(x),cos(2*x)])
#Se grafica el resultado obtenido
figd=plot(d,-pi,pi,color="red", title="Proyección con calc_proy()",legend_label='(d) calc_proy()')

#se utiliza fourier_series_cosine_coefficient() y fourier_series_sin_coefficient() como alternativa para comprobar.
cc0=f.fourier_series_cosine_coefficient(0)
cc1=f.fourier_series_cosine_coefficient(1)
cc2=f.fourier_series_cosine_coefficient(2)
d1=(cc0/2)+cc1*cos(x)+cc2*cos(2*x)
#Se grafica el resultado obtenido
figd1=plot(d1,-pi,pi,color="orange",title="Proyección con coeficientes de fourier",legend_label='(d) coeficientes de fourier')

#Se utiliza graphics_array() para mostrar en dos gráfica los resultados anteriores
figure= graphics_array([figf+figd,figf+figd1])
figure.show(figsize=[8,4])
```

---
$(e) \: span\{1, sin(x), sin(2x), cos(x), cos(2x)\}$

```python
#Se utiliza la función para calcular con el conjunto {1,sin(x),sin(2*x),cos(x),cos(2*x)}
e=calc_proy([1,sin(x),sin(2*x),cos(x),cos(2*x)])
#Se grafica el resultado obtenido
fige=plot(e,-pi,pi,color="red", title="Proyección con calc_proy()",legend_label='(e) calc_proy()')

#se utiliza fourier_series_cosine_coefficient() y fourier_series_sin_coefficient() como alternativa para comprobar.
cc0=f.fourier_series_cosine_coefficient(0)
cs1=f.fourier_series_sine_coefficient(1)
cs2=f.fourier_series_sine_coefficient(2)
cc1=f.fourier_series_cosine_coefficient(1)
cc2=f.fourier_series_cosine_coefficient(2)
e1=(cc0/2)+cs1*sin(x)+cs2*sin(2*x)+cc1*cos(x)+cc2*cos(2*x)
#Se grafica el resultado obtenido
fige1=plot(e1,-pi,pi,color="orange",title="Proyección con coeficientes de fourier",legend_label='(e) coeficientes de fourier')

#Adicionalmente, se puede utilizar fourier_series_partial_sum() para comprobar
e2=f.fourier_series_partial_sum(2)
fige2=plot(e1,-pi,pi,color="purple",title="Proyección con sumas parciales de fourier",legend_label='(e) fourier_series_partial_sum()')

#Se utiliza graphics_array() para mostrar en dos gráfica los resultados anteriores
figure= graphics_array([figf+fige,figf+fige1,figf+fige2])
figure.show(figsize=[10,4])
```

<p align="justify">De lo anterior, se puede observar que realizar la proyección por medio de $(<f,gi>/<gi,gi>)*gi$, teniendo en cuenta que el producto interno está definido por medio de la integral, resulta equivalente a la serie de fourier o es una forma de obtenerla. Además, está claro que entre mayor cantidad de funciones se utilicen para generar, mejor será la aproximación a la función $f$.</p>

Para finalizar, se muestran las diferentes funciones encontradas para compararlas entre sí.

```python
figaf=plot(a,-pi,pi,color="red", title="Comparación",legend_label='(a)')
figbf=plot(b,-pi,pi,color="green", title="Comparación",legend_label='(b)')
figcf=plot(c,-pi,pi,color="brown", title="Comparación",legend_label='(c)')
figdf=plot(d,-pi,pi,color="purple", title="Comparación",legend_label='(d)')
figef=plot(e,-pi,pi,color="black", title="Comparación",legend_label='(e)')

more=f.fourier_series_partial_sum(10)
figmore=plot(more,-pi,pi,color="black", title="Serie de Fourier con n=10")

figure= graphics_array([figf+figaf+figbf+figcf+figdf+figef,figf+figmore])
figure.show(figsize=[10,4])
```

<p align="justify">Como ya se había dicho, al hacer la serie de fourier con $n=10$ se obtiene una mejor aproximación; a medida que se aumenta, va mejorando la aproximación a la función</p>

<p align="justify">Se puede deducir que uno de las funciones que necesariamente deben estar es $1$, ya que se podría decir es la que "centra" la aproximación y no la deja sobre el eje x, teniendo en cuenta que mi función $f$ está un poco más arriba de este. Ahora, para escoger las otras dos funciones, es importante tener en cuenta que las de $n=1$ son las que dan la "forma" inicial a la aproximación, y a medida que se van agregando más senos y cosenos lo que se hace es agregar más detalle; sin embargo, las fundamentales son las de menor $n$.<br>
Con la función $f$ que tengo, sucede algo particular, y es que la función seno es más parecida a $f$ que la función coseno. Por esta razón, resulta mejor escoger $sen(2x)$ en lugar de $cos(x)$.<br>

<p align="justify">Para hacer una pequeña muestra de lo dicho anteriormente, se presenta a continuación una función para calcular la distancia a la función $f$. Con esta, se realizan varias pruebas entre diferentes conjuntos para observar cuál produce la menor distancia. <br>
Al final se muestra que la distancia que hay con la serie de fourier con $n=10$ es mucho menor, debido precisamente al $n$.</p>

```python
def calc_dist(w):
    #Se recorre cada una de las funciones del conjunto generador
    #producto interno coeficiente de Fourier
    distancia=sqrt(integral(((18-x^2)-w)^2,x,-pi,0)+integral(((8*x+4)-w)^2,x,0,pi))
    return distancia
```

```python
opc=[]
opc1=calc_proy([1,sin(x),sin(2*x)])
#Se grafica el resultado obtenido
figopc1=plot(opc1,-pi,pi,color="red", title="Proyección con calc_proy()",legend_label='opc1')
opc.append(opc1)

opc2=calc_proy([1,cos(x),cos(2*x)])
#Se grafica el resultado obtenido
figopc2=plot(opc2,-pi,pi,color="green", title="Proyección con calc_proy()",legend_label='opc2')
opc.append(opc2)

opc3=calc_proy([1,sin(x),cos(x)])
#Se grafica el resultado obtenido
figopc3=plot(opc3,-pi,pi,color="brown", title="Proyección con calc_proy()",legend_label='opc3')
opc.append(opc3)

opc4=calc_proy([1,sin(4*x),sin(5*x)])
#Se grafica el resultado obtenido
figopc4=plot(opc4,-pi,pi,color="purple", title="Proyección con calc_proy()",legend_label='opc4')
opc.append(opc4)

opc5=calc_proy([1,cos(4*x),cos(5*x)])
#Se grafica el resultado obtenido
figopc5=plot(opc5,-pi,pi,color="black", title="Proyección con calc_proy()",legend_label='opc5')
opc.append(opc5)

for i in range(0,5):
    print("Distancia opc"+str(i+1)+": "+str(calc_dist(opc[i]).n()))

print("Distancia serie de fourier con n=10: "+str(calc_dist(more).n()))
    
#Se grafican todos los resultados anteriores
figure= (figf+figopc1+figopc2+figopc3+figopc4+figopc5)
figure.show()
```
