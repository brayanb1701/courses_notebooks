# Series de Tiempo

## Método Clásico

Para mirar desde el punto práctico lo relacionado con la descomposición clásica de series de timepo vamos a similar un conjunto de datos con diferentes estacionalidades a partir de una distribución normal aleatoria


```r
datos <- rnorm(300,mean = 20, sd = 50)
plot(datos,type = 'l')
```

Ahora, esos datos, vamos a convertirlos en serie de tiempo mensual

```r
serie <- ts(datos, start = c(1995,1), frequency = 12)
plot(serie)
```

```r
serie
```

Descompongamos la seie, para ver cómo nos resulta

```r
serie.desc <- decompose(serie)
```

```r
## Gráfico
plot(serie.desc)
```

```r
serie.desc
```

Dijimos que construir una media móvil podría suavizar la serie. Si aplicamos una media móvil de orden 12 (debería ser igual a la frecuencia estacional) veamos cómo cambia la serie. Necesitamos la función `movavg` de la librería `pracma`

```r
install.packages('pracma')
```

```r
library(pracma)
```

```r
serie.m12 <- movavg(serie, 12)
```

```r
plot(datos,type = 'b')
lines(serie.m12, col = 2, type = 'b')
```

```r
s.m12 <- ts(serie.m12,start = c(1995,1), frequency = 12)
plot(s.m12)
```

```r
plot(serie)
```

```r
plot(decompose(s.m12))
```

```r
decompose(s.m12)
```

## Metodología Box-Jenkings

### ¿Cómo Identificamos las Series de Tiempo?

| TIPO |            AR(p)            |            MA(q)            |   ARMA(p,q)  |    ARIMA(p,d,q)    |
|:----:|:---------------------------:|:---------------------------:|:------------:|:------------------:|
|  FAC |         Se desvanece        | Se corta luego del rezago q | Se desvanece | Decrece lentamente |
| FACP | Se corta luego del rezago p |         Se desvanece        | Se desvanece |    Se desvanece    |

Necesitamos la librería `forecast` para poder hacer pronósticos y estimaciones

```r
install.packages('forecast')
```

```r
# Librería para simular series e identificar series de manera automática
library(forecast)
```

Veamos un par de gráficas útiles: función de autocorrelación y función de autocorrelación parcial para ver si podemos identificar el orden del a serie

```r
par(mfrow=c(2,2))
acf(serie)
pacf(serie)
acf(s.m12)
pacf(s.m12)
```

Podemos identificar el orden de la serie con la función `auto.arima` para tener una primera aproximación

```r
# La serie original
auto.arima(serie)
```

```r
# La serie con media móvil
auto.arima(s.m12)
```

Aparece en la estimación de la serie un elemento que no hemos visto en clase, pues el tema es sólo introductorio, se trata de un componente estacional que acompaña un proceso ARIMA. Para reducir ese efecto, podemos hacer uso de una metodología llamada diferenciación, que consiste en calcular la diferencia con un número definido de rezagos, probemos con la primera diferencia y veamos que pasa:

```r
s.diff <- diff(serie)
s.m12.diff <- diff(s.m12)
```

```r
par(mfrow=c(2,2))
acf(s.diff)
pacf(s.diff)
acf(s.m12.diff)
pacf(s.m12.diff)
```

```r
ms1 <- auto.arima(s.diff)
ms1
```

```r
ms2 <- arima(s.diff, order = c(0,0,2))
ms2
```

```r
ms3 <- arima(s.diff, order = c(1,0,2))
ms3
```

```r
AIC(ms1,ms2,ms3)
BIC(ms1,ms2,ms3)
```

```r
args(arima)
```

```r
auto.arima(s.m12.diff)
```

Hagamos un pequeño pronóstico con la serie sin media móvil, usamos la función `forecast`:

```r
forecast(serie,6)
```
