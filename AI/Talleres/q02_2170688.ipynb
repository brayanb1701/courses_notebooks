### Ejercicio 1  <a name="eje1"></a>

En machine learning un factor importante es utilizar conjunto de datos clásicos y evaluar el comportamiento de los clasificadores. Teniendo en cuenta lo anterior, construya una función que: 

1. Cargue el dataset iris(), disponible en la libreria  `sklearn`
2. Cree dos clasificadores siguiendo la siguiente configuración: 
    1. Una máquina de soporte vectorial (SVC) con un kernel polinomial (poly) de grado 7, y `gamma='auto'`
    2. Un RandomForestClassifier con 100 estimadores y una profundidad máxima de 10 niveles
4. Clasifique el dataset con los dos clasificadores configurados.
    - Utilice una única partición de los datos, tomando el 35% como test
5. La función debe retornar los objetos SVM y RF entrenados. También una lista de los score obtenidos de la forma: `[score_rf, score_sv]`

Referencia SVC: [ https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html ] <br>
Referencia RandomForestClassifier: [ https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html ]

```python
def fun_q1():
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.svm import SVC
    from sklearn.model_selection import train_test_split
    from sklearn.datasets import load_iris
    import numpy as np
    
    iris = load_iris()
    X = iris.data
    y = iris.target

    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.35)
    a = SVC(kernel='poly', degree=7, gamma='auto')
    b=RandomForestClassifier(n_estimators=100, max_depth=10)
    a.fit(X_train,y_train)
    b.fit(X_train,y_train)
    c=a.score(X_test,y_test)
    d=b.score(X_test,y_test)
    return a, b, [c, d]

#import urllib, inspect
#src1 = urllib.quote_plus(inspect.getsource(fun_q1))
import urllib.parse, inspect
src1 = urllib.parse.quote_plus(inspect.getsource(fun_q1))
```

Completa la siguiente función para que realice una regresión utilizando un  **DecisionTreeRegressor** pasando como argumento **X_train** &nbsp; e &nbsp; **y_train**.<br>La función debe devolver el **objeto del regresor** y el **resultado** usando la siguiente configuración: 


- El regresor `DecisionTreeRegressor` debe ser entrenado con los parametros de `max_depth=8` y `criterion='mse'`
- La evaluación del método debe ser exhaustiva, utilizando un esquema de *cross-validation* con un **KFold = 20**
- La metrica para la evaluación es el MSE. Recuerde que en sklearn, se cuenta con la métrica MSE. 
- Se debe retornar el valor promedio de la métrica MSE. 
- se debe retornar el estimador ya entrenado

Referencia DecisionTreeRegressor [ https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html ]

```python
def fit_reg(X_train, y_train):
    import numpy as np
    import pandas as pd
    from sklearn.tree import DecisionTreeRegressor
    from sklearn.model_selection import cross_val_score
    from sklearn.model_selection import KFold
    from sklearn.metrics import make_scorer, mean_squared_error

    est=DecisionTreeRegressor(max_depth=8, criterion='mse')
    s = cross_val_score(est, X_train, y_train, cv=KFold(20, shuffle=True), scoring=make_scorer(mean_squared_error))
    score=np.mean(s)
    return est, score

#import urllib, inspect
#src6 = urllib.quote_plus(inspect.getsource(fit_reg))
import urllib.parse, inspect
src6 = urllib.parse.quote_plus(inspect.getsource(fit_reg))
```

### Ejercicio 4  <a name="eje4"></a>

Deep Learning con dataset **Fashion-MNIST**.

<!--<img src="../../imgs/embedding.gif" style="width:300px;">-->

- Cargue los datos desde el archivo `"data/fashion2.csv"`
- Tome el  90%  del dataset como training y el restante para el test
- Realice la clasificación con: 
    - DNN con la siguiente arquitectura
        - capa de entrada que reciba el vector de 784
        - Una capa oculta densa de 512 neuronas y activación ReLu
        - Una capa oculta densa de 128 neuronas y activación ReLu
        - Una capa de salida, según las clases del dataset. 
        - Entrene el modelo con la configuración: `epochs=5, batch_size=1000`
- La función debe devolver el modelo `DNN` y el resultado `score_DNN`

```python
def clas_fashion2():
    import tensorflow as tf
    from tensorflow import keras
    import numpy as np
    import pandas as pd
    from sklearn.model_selection import train_test_split

    d = pd.read_csv("data/fashion2.csv",header=None)
    X = d.values[:,1:]
    y = d.values[:,0]
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.10)
    model = keras.models.Sequential([
    keras.layers.Flatten( input_shape=(X_train[0].shape)),
    keras.layers.Dense(512, activation=tf.nn.relu),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
    ])

    model.compile(optimizer=tf.optimizers.Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
    
    model.fit(X_train, y_train, epochs=5, batch_size=1000, verbose=0)
    test_loss, test_acc = model.evaluate(X_test, y_test)
   
    return model, test_acc

#import urllib, inspect
#src3 = urllib.quote_plus(inspect.getsource(clas_fashion2))
import urllib.parse, inspect
src3 = urllib.parse.quote_plus(inspect.getsource(clas_fashion2))
```
