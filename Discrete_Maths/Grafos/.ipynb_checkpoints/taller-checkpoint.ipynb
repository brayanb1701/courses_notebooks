<h3 align="center"> Solución del inciso #1:</h3><br> 

Para solucionar el primer punto, implementamos las librerías de networkx y matplotlib, las cuales se deben instalar para su ejecución; utilizamos un bloc de notas (graph.gml) donde colocamos las aristas con sus respectivos pesos.



```python
#ESTA ES LA SOLUCIÓN DE LA REPRESENTACIÓN GRÁFICA DEL GRAFO

import matplotlib.pyplot as plt #Importamos a la librería matplotlib (plt)
import networkx as nx #Importamos a la librería networkx (nx)

G = nx.read_gml('graph.gml') #Llamamos al archivo GML que contiene la información del grafo y se lo asignamos a la variable G
# Colocamos las posiciones de los vertices
pos = {"A": (2, 1),
       "B": (19, 1),
       "C": (5, 2),
       "D": (11, 3),
       "E": (18, 5),
       "F": (4, 6),
       "G": (12, 7),
       "H": (20, 8),
       "I": (16, 10),
       "J": (10, 10),
       "K": (7, 11),
       "L": (5, 13),
       "M": (11, 13),
       "N": (19, 14),
       "O": (16, 16),
       "P": (4, 17),
       "Q": (9, 17),
       "R": (11, 19),
       "S": (1, 20),
       "T": (20, 20)}
#Dibujo del grafo
nx.draw_networkx_nodes(G, pos, ax=plt.axes((4, 4+.25*2, 3, 3)), node_size=3000,  node_color='g') #Le damos un tamaño al grafo y un color
nx.draw_networkx_labels(G, pos, font_weight='bold', font_size=24) #En los nodos le asignamos un tamaño a los vertices
nx.draw_networkx_edges(G, pos, alpha=1, width=2) #Dibujamos las aristas
nx.draw_networkx_edge_labels(G, pos, font_size=14) #Dibujamos los pesos de las aristas
plt.axis("off") 

plt.show() #Decimos que muestre el grafo

```

    dist = {}                                                    //Diccionario de distancias finales
    seen = {}                                                    
    c = count()                                                  //La variable c se utiliza para evitar comparar nodo
    fringe = []                                                  //fringe es un heap con tres tuplas (distancia, c , nodo)

    while fringe:                                                //Mientras haya elementos en el heap

                continue

            if cutoff is not None:                               
                if vu_dist > cutoff:                             
                    continue

                if vu_dist < dist[u]:                            //Verifica que no hayan patrones contradictorios
