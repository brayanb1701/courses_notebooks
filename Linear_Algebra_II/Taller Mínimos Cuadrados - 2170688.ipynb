Puntos: $P1 = (1, 0)$, $P2 = (2, 12)$, $P3 = (4, 15)$ y $P4 = (5,8)$ 

- Buscar la mejor curva de tipo:<br>
    $y = ax^2 + bx + c$<br>
    ¡Una parábola! que mejor se aproxime a P1, P2, P3, P4.

Buscamos $\hat{x} = (Q^{T}Q)^{−1}Q^{T}b$; la proyección es: $Q\hat{x}$

```python
p1=(1,0) ; p2=(2,12) ; p3=(4,15) ; p4=(5,8)
```

```python
M=matrix(QQ,[[1,1,1],[4,2,1],[16,4,1],[25,5,1]])
Mt=M.transpose()
Mp=Mt*M
b=vector([0,12,15,8])
x=Mp.inverse()*Mt*b
print(x,M*x)
```

```python
g1=Graphics()
f(x)=(-19/6)*x^2+(209/10)*x-(263/15)
g1+=points([p1 , p2, p3, p4],color='red',size=20)
g1+=plot(f,[0,8])
g1.show()
```

- Debes buscar la mejor curva de tipo:<br>
  $x^2 + y^2 = ax + by + d$<br>
  ¡Un círculo! que mejor se aproxime a P1, P2, P3, P4.

Buscamos $\hat{x} = (Q^{T}Q)^{−1}Q^{T}b$; la proyección es: $Q\hat{x}$

```python
M2=matrix(QQ,[[1,0,1],[2,12,1],[4,15,1],[5,8,1]])
Mt2=M.transpose()
Mp2=Mt2*M2
b2=vector([1,148,241,89])
x2=Mp2.inverse()*Mt2*b2
print(x2,M2*x2)
```

```python
g2=Graphics()
h(x,y)=x^2+y^2-((-4455/1813)*x+(28013/1813)*y-(-14642/1813))
g2+=points([p1 , p2, p3, p4],color='red',size=20)
g2+=implicit_plot(h,(-10,8),(-2,17),color='magenta')
g2.show()
```

- Compara los dos resultados.

Solamente por inspección gráfica, ya se observan las diferencias entre ambas aproximaciones, siendo la parábola una aproximación mucho mejor que el círculo. Además, se pueden comparar las distancias de los vectores obtenidos para determinar la precisión.

```python
bp=vector([0,12,15,8])
xp=M*x
distancia1=(bp-xp).norm()
print(distancia1.n())

bc=vector([1,148,241,89])
xc=M2*x2
distancia2=(bc-xc).norm()
print(distancia2.n())
```

Se evidencia la notoria diferencia entre las distancias de los respectivos vectores generados por las aproximaciones de la parábola y el círculo, siendo mucho mejor la primera de estas.
